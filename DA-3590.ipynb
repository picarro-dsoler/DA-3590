```python
import sqlite3
import pandas as pd
import os

# This will work in notebooks/Jupyter (where __file__ is not defined) by using the current working directory
db_path = os.path.join(
    os.getcwd(), "database", "italgas_g2g.db"
)

if not os.path.exists(db_path):
    raise FileNotFoundError(f"Database file not found at: {db_path}")

# Get the names of all tables in the database
with sqlite3.connect(db_path) as con:
    tables_df = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", con)
    table_names = tables_df['name'].tolist()

    print("Tables in database:", table_names)

with sqlite3.connect(db_path) as con:
    # Get the names of all tables in the database
    tables_df = pd.read_sql_query("SELECT * FROM Leaks", con)
tables_df.columns
```
